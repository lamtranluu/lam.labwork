# Keywords extractions using BOW feature

## DC

```python
# Create document term matrix for DC - BoW
vectorizerDC = CountVectorizer(analyzer='word', min_df=0.0, max_df=1.0, ngram_range=(2, 2))
DC_vector = vectorizerDC.fit_transform(dc['content_cleaned'])

#DC-Positive comments:
vectorizerDC_pos = CountVectorizer(analyzer='word', min_df=0.0, max_df=1.0, ngram_range=(2, 2))
DC_vector_pos = vectorizerDC_pos.fit_transform(dc['content_cleaned'][dc["sentiments"]=="Positive"])

#DC-Negative comments:
vectorizerDC_neg = CountVectorizer(analyzer='word', min_df=0.0, max_df=1.0, ngram_range=(2, 2))
DC_vector_neg = vectorizerDC_neg.fit_transform(dc['content_cleaned'][dc["sentiments"]=="Negative"])
```

```python
DC_pos_words=pd.DataFrame(words_freq_DC_pos,columns=['words','num'])
```

# Extract negative, positive reviews based on each topic

```python
def get_keyword_scores(keywords):
    df = pd.DataFrame(columns = ["keyword","Marvel_pos_counter","Marvel_neg_counter","Marvel_pos_tfidf","Marvel_neg_tfidf","DC_pos_counter","DC_neg_counter","DC_pos_tfidf","DC_neg_tfidf"])
    for i in range(len(keywords)):
        df.at[i,"keyword"] = keywords[i]
        df.at[i,"Marvel_pos_counter"] = Marvel_vector_pos[:,vectorizerMarvel_pos.vocabulary_[keywords[i]]].sum()
        df.at[i,"Marvel_neg_counter"] = Marvel_vector_neg[:,vectorizerMarvel_neg.vocabulary_[keywords[i]]].sum()
        df.at[i,"DC_pos_counter"] = DC_vector_pos[:,vectorizerDC_pos.vocabulary_[keywords[i]]].sum()
        df.at[i,"DC_neg_counter"] = DC_vector_neg[:,vectorizerDC_neg.vocabulary_[keywords[i]]].sum()
        
    return df
```
